Operations which give the settingwithcopywarning 


```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["John", "Alice", "Bob", "Emma"],
    "Age": [25, 17, 30, 16],
    "Marks": [85, 92, 78, 88]
})

# Filter students who are adults
adults = df[df["Age"] >= 18]

# Warning occurs here
adults["Status"] = "Adult"
```

```python
print(adults)
```

```python
print(df)  # Original DataFrame remains unchanged
```

```python
# Filter students who are adults
adults = df[df["Age"] >= 18]

# Warning occurs here
adults["Status"] = "[Adult]"
```

```python
print(adults)
```

```python
# Filter students who are adults
adults = df[df["Age"] >= 18].copy()

# Warning occurs here
adults["Status"] = "[Adult]"
```

```python
print(adults)
```

```python
print(df)
```

```python
import pandas as pd

sales = pd.DataFrame({
    "Amount": [500, 1500, 2500, 800, 3200]
})

def discount(amount):
    if amount >= 2000:
        return amount * 0.20
    elif amount >= 1000:
        return amount * 0.10
    else:
        return 0

sales["Discount"] = sales["Amount"].apply(discount)

print(sales)
```

```python
import numpy as np

conditions = [
    sales["Amount"] >= 2000,
    sales["Amount"] >= 1000
]

choices = [
    sales["Amount"] * 0.20,
    sales["Amount"] * 0.10
]

sales["Discount"] = np.select(conditions, choices, default=0)

print(sales)
```
