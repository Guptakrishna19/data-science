```python
import pandas as pd
import numpy as np
```

```python
data = {
    'OrderID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Customer': ['Amit', 'Priya', 'Rahul', np.nan, 'Sneha', 'Amit', 'Rahul', 'Priya'],
    'City': ['Mumbai', 'Delhi', 'Mumbai', 'Pune', np.nan, 'Mumbai', 'Mumbai', 'Delhi'],
    'OrderDate': ['2024-01-05', '2024-01-15', '2024-02-01', '2024-02-20',
                  '2024-03-10', '2024-03-15', '2024-04-01', '2024-04-10'],
    'Amount': [1200, np.nan, 800, 1500, 950, 1100, np.nan, 2000],
    'Quantity': [2, 1, 4, 3, 2, 1, 5, 2]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Q1.1
print(df.isna().sum())
```

```python
# Q1.2
df['Amount'] = df['Amount'].fillna(df['Amount'].mean())

print(df)
```

```python
# Q1.3
df['City'] = df['City'].fillna('Unknown')
print(df)
```

```python

# Q1.4
df = df.dropna(subset=['Customer']).reset_index(drop=True)
print(df)
```

Exercise 2 map , apply 

```python
# Q2.1
city_map = {'Mumbai': 'MUM', 'Delhi': 'DEL', 'Pune': 'PUN', 'Unknown': 'UNK'}
df['CityCode'] = df['City'].map(city_map)
print(df)
```

```python
# Q2.2
df['OrderSize'] = df['Quantity'].apply(lambda x: 'Bulk' if x > 3 else 'Normal')
print(df)
```

```python
# Q2.3
df['UnitPrice'] = df.apply(lambda row: row['Amount'] / row['Quantity'], axis=1)
print(df)
```

exercise 3 datetime

```python
# Q3.1
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
print(df)
```

```python

# Q3.2

df['Month'] = df['OrderDate'].dt.month
df['Weekday'] = df['OrderDate'].dt.day_name()
print(df)
```

```python

# Q3.3
march_orders = df[(df['OrderDate'].dt.month == 3) & (df['OrderDate'].dt.year == 2024)]
print(march_orders)
```

pivot table

```python
# Q4.1
pivot1 = pd.pivot_table(df, values='Amount', index='City', aggfunc='sum')
print(pivot1)
```

```python
# Q4.2
pivot2 = pd.pivot_table(df, values='Quantity', index='City', columns='Month', aggfunc='mean')
print(pivot2)
```

```python
# Q4.3
pivot3 = pd.pivot_table(df, values='Amount', index='City', aggfunc='sum', margins=True, margins_name='Total')
print(pivot3)
```

method chainning 

```python
result = (
    df
    .assign(Amount=lambda d: d['Amount'].fillna(d['Amount'].mean()))
    .assign(OrderSize=lambda d: d['Quantity'].apply(lambda x: 'Bulk' if x > 3 else 'Normal'))
    .assign(OrderDate=lambda d: pd.to_datetime(d['OrderDate']))
    .query("City == 'Mumbai'")
)
print(result)
```
