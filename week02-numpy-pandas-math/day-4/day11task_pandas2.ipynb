```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("messy_dataset.csv")
```

```python
df.head()
```

missing values

```python
df.isna().sum()
```

removing duplicates

```python
duplicates = df.duplicated().sum()

print("Duplicate rows:", duplicates)

df = df.drop_duplicates().reset_index(drop=True)
```

```python
df["Customer"] = df["Customer"].str.strip()

df["City"] = df["City"].str.strip()
```

```python
df["City"] = df["City"].str.title()

df["Customer"] = df["Customer"].str.title()
```

missing names 

```python
df = df.dropna(subset=["Customer"]).reset_index(drop=True)
```

```python
median_age = df["Age"].median()

df["Age"] = df["Age"].fillna(median_age)
```

```python
mode_city = df["City"].mode()[0]

df["City"] = df["City"].fillna(mode_city)
```

```python
df["OrderDate"] = pd.to_datetime(
    df["OrderDate"],
    errors="coerce"
)
```

```python
df = df.dropna(subset=["OrderDate"]).reset_index(drop=True)
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
df["Price"] = df["Price"].astype(float)
```

```python
df = df[df["Price"] > 0].reset_index(drop=True)
```

```python
df = df[df["Age"] > 0].reset_index(drop=True)
```

```python
pivot = df.pivot_table(
    values="Price",
    index="City",
    aggfunc="mean"
)

pivot
```

```python
pivot = df.pivot_table(
    values="Price",
    index="City",
    columns="Gender",
    aggfunc="sum"
)

pivot
```

```python
df.to_csv(
    "cleaned_dataset.csv",
    index=False
)
```

| Cleaning Decision                     | WHY                                                  |
| ------------------------------------- | ---------------------------------------------------- |
| Removed duplicate rows                | Prevent duplicate records from affecting analysis    |
| Removed rows with missing Customer    | Customer name is an important identifier             |
| Filled missing Age using median       | Median is less affected by outliers                  |
| Filled missing City using mode        | Most common city is a reasonable replacement         |
| Trimmed spaces from Customer and City | Prevents duplicate categories caused by extra spaces |
| Converted City names to title case    | Makes categories consistent                          |
| Converted OrderDate to datetime       | Required for date analysis                           |
| Removed invalid dates                 | Invalid dates cannot be analyzed                     |
| Removed negative Price values         | Product prices cannot be negative                    |
| Removed invalid Age values            | Age cannot be zero or negative                       |
| Extracted Year, Month and Weekday     | Enables time-based analysis                          |
| Reset index after row removal         | Maintains a clean sequential index                   |
