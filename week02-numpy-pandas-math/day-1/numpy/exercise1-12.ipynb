boolean mask 

```python
arr = np.arange(10)
mask = arr > 5
print(mask)
print(arr[mask])
```

Multiple condition 

```python
arr = np.arange(20)
result = (arr > 5) & (arr < 15)
result2 = arr[(arr > 5) & (arr < 15)]
print(result)
print(result2)
```

Axis aggregation

```python
matrix = np.arange(12).reshape(3,4)

print(matrix.sum())

print(matrix.sum(axis=0))

print(matrix.sum(axis=1))
```

Statistics

```python
arr = np.array([5,10,15,20,25])

print(np.mean(arr))
print(np.median(arr))
print(np.max(arr))
print(np.min(arr))
print(np.std(arr))
```

Broadcasting

```python
arr = np.array([1,2,3,4])

result = arr + 10

print(result)
```

Broadcasting with matrix


```python
matrix = np.array([
    [1,2,3],
    [4,5,6]
])

vector = np.array([10,20,30])

result = matrix + vector

print(result)
```

Element-wise operations

```python
a = np.array([1,2,3])

b = np.array([4,5,6])

print(a+b)

print(a*b)

print(a**2)

print(np.sqrt(a))
```

Python loop version

```python
numbers = list(range(1000000))

squared = []

for n in numbers:
    squared.append(n * n)
```

numpy version

```python
arr = np.arange(1000000)

squared = arr * arr
```

## Benchmark Note

I compared squaring one million numbers using a Python loop and a NumPy vectorized operation.

The Python loop took significantly longer because each iteration is executed individually by the Python interpreter.

The NumPy solution performed the operation on the entire array at once using optimized C-level implementations.

The vectorized NumPy operation was approximately 30–50 times faster on my machine.

This demonstrates why NumPy is preferred for numerical computing and large-scale data processing.
