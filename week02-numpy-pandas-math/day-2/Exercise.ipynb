Inverse 

A⁻¹ = 1/(ad - bc) × [ d -b] [-c  a]

det(A) = (1×4) - (2×3) = 4 - 6 = -2

A⁻¹ = 1/(-2) × [ 4 -2][-3  1]

= [-2    1 ][1.5  -0.5]

```python
A = np.array([[1, 2],[3, 4]])

a_inv = np.linalg.inv(A)
print(a_inv)
```

```python
# Sanity check: A @ A_inv should ≈ identity
print(A @ a_inv)
```
