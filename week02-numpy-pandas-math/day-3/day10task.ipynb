```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv")
```

```python
sales.info()
```

```python
sales.describe()
```

1 total revenue

```python
sales["Sales"].sum()
```

2 total profit

```python
sales["Profit"].sum()
```

3 average value

```python
sales["Sales"].mean()
```

4 top 10 products

```python
sales.groupby("Product")["Sales"] \
.sum() \
.sort_values(ascending=False) \
.head(10)
```

5 revenue by region

```python
sales.groupby("Region")["Sales"].sum()
```

6 Profit by region

```python
sales.groupby("Region")["Profit"].sum()
```

7 best category

```python
sales.groupby("Category")["Sales"] \
.sum() \
.sort_values(ascending=False)
```

8 best sub - category 

```python
sales.groupby("Sub-Category")["Sales"] \
.sum() \
.sort_values(ascending=False)
```

9 higest selling customer

```python
sales.groupby("Customer Name")["Sales"] \
.sum() \
.sort_values(ascending=False) \
.head()
```

10 average discount 

```python
sales["Discount"].mean()
```

11 most orderd product

```python
sales["Product"].value_counts().head()
```

12 higest profit product

```python
sales.groupby("Product")["Profit"] \
.sum() \
.sort_values(ascending=False) \
.head()
```

13 top 5 cities

```python
sales.groupby("City")["Sales"] \
.sum() \
.sort_values(ascending=False) \
.head()
```

14 region wise statistics

```python
sales.groupby("Region").agg({
    "Sales":["sum","mean","max"],
    "Profit":["sum","mean"],
    "Discount":"mean"
})
```
