# Reflection

### What I Learned
- Gained hands-on experience with Pandas for data selection, filtering, sorting, grouping, and aggregation.
- Learned the practical difference between `loc` and `iloc` for accessing data.
- Used boolean filtering to extract meaningful subsets of data based on business conditions.
- Applied `groupby()` and `agg()` to summarize sales, profit, and other key business metrics.
- Practiced sorting data with `sort_values()` and combining datasets using `merge()`.

### Challenges Faced
- Choosing between `loc` and `iloc` in different scenarios.
- Writing multiple aggregation functions with `groupby().agg()`.
- Ensuring filters with multiple conditions returned the expected results.

### How I Solved Them
- Referred to the Pandas documentation and practiced with small examples.
- Verified intermediate outputs after each operation.
- Cleaned the dataset by handling missing values and removing duplicates before analysis.

### Areas to Improve

- Explore advanced merge techniques and time-series analysis.
- Improve efficiency by chaining Pandas operations where appropriate.
