```python
import duckdb                             

conn = duckdb.connect("business.db")
```

```python
conn.execute("""
CREATE OR REPLACE TABLE customers(
    customer_id INTEGER,
    customer_name VARCHAR,
    city VARCHAR
);
""")

conn.execute("""
INSERT INTO customers VALUES
(1,'Krish','Mumbai'),
(2,'Rahul','Pune'),
(3,'Aman','Delhi'),
(4,'Priya','Mumbai'),
(5,'Neha','Bangalore');
""")
```

```python
conn.execute("""
CREATE OR REPLACE TABLE products(
    product_id INTEGER,
    product_name VARCHAR,
    category VARCHAR,
    price DOUBLE
);
""")

conn.execute("""
INSERT INTO products VALUES
(1,'Laptop','Electronics',60000),
(2,'Phone','Electronics',30000),
(3,'Mouse','Accessories',1000),
(4,'Keyboard','Accessories',2000),
(5,'Monitor','Electronics',15000);
""")
```

```python
conn.execute("""
CREATE OR REPLACE TABLE orders(
    order_id INTEGER,
    customer_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    order_date DATE
);
""")

conn.execute("""
INSERT INTO orders VALUES
(101,1,1,1,'2025-01-01'),
(102,2,2,2,'2025-01-02'),
(103,1,3,5,'2025-01-03'),
(104,3,1,1,'2025-01-04'),
(105,4,5,2,'2025-01-05'),
(106,5,2,1,'2025-01-06');
""")
```

select

```python
conn.sql("""
SELECT *
FROM customers;
""").df()
```

where

```python
conn.sql("""
SELECT *
FROM customers
WHERE city='Mumbai';
""").df()
```

Group by 

```python
conn.sql("""
SELECT
city,
COUNT(*) AS total_customers
FROM customers
GROUP BY city;
""").df()
```

having 

```python
conn.sql("""
SELECT
city,
COUNT(*) AS total_customers
FROM customers
GROUP BY city
HAVING COUNT(*) > 1;
""").df()
```

join

```python
conn.sql("""
SELECT
c.customer_name,
o.order_id
FROM customers c
JOIN orders o
ON c.customer_id=o.customer_id;
""").df()
```

subquery

```python
conn.sql("""
SELECT *
FROM products
WHERE price >
(
SELECT AVG(price)
FROM products
);
""").df()
```

window function

```python
conn.sql("""
SELECT
product_name,
price,
RANK() OVER(
ORDER BY price DESC
) AS rank_num
FROM products;
""").df()
```
