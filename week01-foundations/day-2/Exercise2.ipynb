```python
numbers = [1, 2, 3, 4]

squares = []

for num in numbers:
    squares.append(num ** 2)

print(squares)
```

Comprehension

```python
numbers = [1, 2, 3, 4]

squares = [num ** 2 for num in numbers]

print(squares)
```

```python
names = ["krish", "rahul", "aman"]

capitalized = [name.upper() for name in names]

print(capitalized)
```

```python
fruits = ["Apple", "Banana", "Orange"]

for i in range(len(fruits)):
    print(i, fruits[i])
```

```python
names = ["Krish", "Rahul", "Aman", "Riya", "Jasi", "Hasti", "Dhudi"]
scores = [90, 85, 95, 88, 92, 80, 78]
grade = ["A", "B", "A", "B", "A", "C", "C"]

for name, score, grade in zip(names, scores, grade):
    print(name, score, grade)
```

```python
def print_details(**details):
    for key, value in details.items():
        print(key, value)

print_details(name="Krish", age=21)
```
