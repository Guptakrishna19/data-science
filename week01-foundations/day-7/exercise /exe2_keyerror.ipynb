```python
shop = {"apple": 30, "banana": 10, "mango": 50}
cart = ["apple", "grapes", "banana", "pineapple"]

total = 0
missing = []

for item in cart:
    if item in shop:
        total += shop[item]
    else:
        missing.append(item)   # track missing items

print("Total:", total)         # 40
print("Not found:", missing)   # ['grapes', 'pineapple']

       # 40
```
