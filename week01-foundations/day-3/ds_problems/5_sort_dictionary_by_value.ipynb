```python
d={
 'a':3,
 'b':1,
 'c':2
}
def sort_by_value(d):

    return dict(
        sorted(
            d.items(),
            key=lambda item:item[1]
        )
    )

print (sort_by_value(d))
```

```python
d={
 'a':3,
 'b':1,
 'c':2,
 'd':0,
}
def sort_by_value(d):

    return dict(
        sorted(
            d.items(),
            key=lambda item:item[1]
        )
    )

print (sort_by_value(d))
```

Complexity
Time - O(n log n)
Space - O(n)
