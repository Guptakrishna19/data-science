normal case 

```python
a=[1, 2, 3, 4]
b=[3, 4, 5, 6]

def intersection(a,b):
    return list(set(a) & set(b))

print(intersection(a, b))
```

boundary case 

```python
a=[1, 2, 9, 7]
b=[3, 4, 5, 6]

def intersection(a,b):
    return list(set(a) & set(b))

print(intersection(a, b))
```
