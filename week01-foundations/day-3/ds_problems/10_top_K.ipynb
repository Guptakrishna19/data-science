normal 

```python
nums = [1,1,1,2,2,3]
k = 2
from collections import Counter

def top_k_frequent(nums,k):

    counts = Counter(nums)

    return [
        item[0]
        for item in counts.most_common(k)
    ]

print(top_k_frequent(nums,k))
```
