```python
words = ["apple", "banana", "grape", "avocado", "blueberry"]
def group_words(words):

    result = {}

    for word in words:

        key = word[0]

        if key not in result:
            result[key] = []

        result[key].append(word)

    return result

print (group_words(words))
```

Complexity
Time - O(n)
Space - O(n)
