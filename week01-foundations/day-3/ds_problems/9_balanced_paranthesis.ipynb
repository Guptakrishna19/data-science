normal case 

```python
def is_balanced(s):
    stack = []

    mapping = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for ch in s:

        if ch in "([{":
            stack.append(ch)

        else:
            if not stack:
                return False

            if stack.pop() != mapping[ch]:
                return False

    return len(stack) == 0
print(is_balanced("({[]})"))
```

error case 

```python
def is_balanced(s):
    stack = []

    mapping = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for ch in s:

        if ch in "([{":
            stack.append(ch)

        else:
            if not stack:
                return False

            if stack.pop() != mapping[ch]:
                return False

    return len(stack) == 0
print(is_balanced("({[()]]})"))
```

Complexity
Time - O(n)
Space - O(n)
