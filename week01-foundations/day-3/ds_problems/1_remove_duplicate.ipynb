normal case 

```python
num=[1, 2, 2, 3, 4, 4, 5, 5, 5, 6]

unique_value = set(num) 

print (list(unique_value))
```

repeated case 

```python
num=[1, 1, 1, 1, 1, 1]

unique_value = set(num) 

print (list(unique_value))
```

when tere is empty list 

```python
num=[]

unique_value = set(num) 

print (list(unique_value))
```

Complexity
Time - O(n)
Space - O(n)
