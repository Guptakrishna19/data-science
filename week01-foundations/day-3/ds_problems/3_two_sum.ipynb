```python
nums=[2, 7, 11, 15, 20, 3, 5]
target = 12

def two_sum(nums,target):

    seen = {}

    for i,num in enumerate(nums):

        diff = target - num

        if diff in seen:
            return [seen[diff], i]

        seen[num] = i

    return []

print (two_sum(nums,target))
```
