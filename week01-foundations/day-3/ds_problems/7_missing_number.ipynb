```python
def missing_number(nums):
    n = len(nums)

    expected = n * (n + 1) // 2

    actual = sum(nums)

    return expected - actual


nums = [3, 0, 1, 2, 5]

result = missing_number(nums)

print("Missing number:", result)
```
